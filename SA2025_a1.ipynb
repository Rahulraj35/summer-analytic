### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('Cars.csv')


avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.lineplot(
    data=avg_mpg_by_year_origin,
    x='model_year',
    y='mpg',
    hue='origin',
    marker='o'
)


plt.title('Average MPG Over the Years by Origin', fontsize=16)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()

```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('Cars.csv')


hue_order = ['japan', 'europe', 'usa']


plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(30, 300),
    alpha=0.7,
    hue_order=hue_order,
    palette='muted'
)


plt.title('Horsepower vs Weight of Cars (Size represents MPG)', fontsize=16)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()

```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd


df = pd.read_csv('Cars.csv')


multi_year_names = df.groupby('name')['model_year'].nunique()
multi_year_names = multi_year_names[multi_year_names > 1].index


filtered_df = df[df['name'].isin(multi_year_names)]


mpg_stats = filtered_df.groupby('name').agg(
    appearances=('model_year', 'count'),
    unique_years=('model_year', 'nunique'),
    std_mpg=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
).reset_index()


consistent_cars = mpg_stats[mpg_stats['std_mpg'] < 1.0]


min_std = consistent_cars['std_mpg'].min()
most_consistent = consistent_cars[consistent_cars['std_mpg'] == min_std]


result = most_consistent.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])


print(result[['name', 'appearances', 'avg_mpg']])

```
